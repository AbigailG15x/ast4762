```python
print('Abigail Glover')
print('PL1 - Mech Systems')
print('DUE Sept 5 at 11:59 pm')
print('**********************')

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import skew, kurtosis, norm
import matplotlib.mlab as mlab

# Load the data from 'PL1data.xlsx'
file_path = 'PL1data.xlsx'
df = pd.read_excel(file_path)

# Label columns according to data labels in excel file
column_x = 'x_i'
column_y = 'y_i'
column_z = 'z_i'

# Create histograms for 'x', 'y', and 'z'
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.hist(df[column_x], bins=20, edgecolor='k')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.title('Histogram of x')

plt.subplot(1, 3, 2)
plt.hist(df[column_y], bins=20, edgecolor='k')
plt.xlabel('Y')
plt.ylabel('Frequency')
plt.title('Histogram of y')

plt.subplot(1, 3, 3)
plt.hist(df[column_z], bins=20, edgecolor='k')
plt.xlabel('Z')
plt.ylabel('Frequency')
plt.title('Histogram of z')

# Ensure proper spacing between subplots
plt.tight_layout()  

# Calculate and print the mean values
mean_x = np.mean(df[column_x])
mean_y = np.mean(df[column_y])
mean_z = np.mean(df[column_z])

print(f"Mean value for 'x': {mean_x}")
print(f"Mean value for 'y': {mean_y}")
print(f"Mean value for 'z': {mean_z}")

# Calculate and print the variance and standard deviation
variance_x = np.var(df[column_x])
variance_y = np.var(df[column_y])
variance_z = np.var(df[column_z])

std_dev_x = np.std(df[column_x])
std_dev_y = np.std(df[column_y])
std_dev_z = np.std(df[column_z])

print(f"\n\nVariance for 'x': {variance_x}")
print(f"Variance for 'y': {variance_y}")
print(f"Variance for 'z': {variance_z}")

print(f"\n\nStandard Deviation for 'x': {std_dev_x}")
print(f"Standard Deviation for 'y': {std_dev_y}")
print(f"Standard Deviation for 'z': {std_dev_z}")

# Calculate and print the range
min_x, max_x = np.min(df[column_x]), np.max(df[column_x])
min_y, max_y = np.min(df[column_y]), np.max(df[column_y])
min_z, max_z = np.min(df[column_z]), np.max(df[column_z])

print(f"\n\nRange for 'x': ({min_x}, {max_x})")
print(f"Range for 'y': ({min_y}, {max_y})")
print(f"Range for 'z': ({min_z}, {max_z})")

# Calculate and print the skewness and excess kurtosis
skewness_x = skew(df[column_x])
skewness_y = skew(df[column_y])
skewness_z = skew(df[column_z])

kurtosis_x = kurtosis(df[column_x], fisher=False)
kurtosis_y = kurtosis(df[column_y], fisher=False)
kurtosis_z = kurtosis(df[column_z], fisher=False)

print(f"\n\nSkewness for 'x': {skewness_x}")
print(f"Skewness for 'y': {skewness_y}")
print(f"Skewness for 'z': {skewness_z}")

excess_kurt_x = kurtosis_x - 3
excess_kurt_y = kurtosis_y - 3
excess_kurt_z = kurtosis_z - 3

print(f"\n\nExcess Kurtosis for 'x':", excess_kurt_x)
print(f"Excess Kurtosis for 'y':", excess_kurt_y)
print(f"Excess Kurtosis for 'z':", excess_kurt_z)

# Spacing
print('\n\n')

# Save the histograms as .png files
plt.savefig('PL1_abigailglover_graph1.png')

# Show the histograms
plt.show()
```

```python

# Load the data from 'PL1data.xlsx'
file_path = 'PL1data.xlsx'
df = pd.read_excel(file_path)

# Replace 'ColumnX', 'ColumnY', and 'ColumnZ' with the actual column names
# from your Excel file
column_x = 'x_i'
column_y = 'y_i'
column_z = 'z_i'

# Define a function to fit and plot a histogram with a Gaussian curve
def plot_histogram_with_gaussian(data, column_name):
    # Fit a Gaussian (normal) distribution to the data
    mu, sigma = norm.fit(data)
    
    # Create a histogram of the data
    n, bins, patches = plt.hist(data, bins=20, density=True, alpha=0.75, facecolor='blue')

    # Calculate the PDF (probability density function) for the Gaussian fit with more points
    x = np.linspace(min(data), max(data), 1000)
    pdf = norm.pdf(x, mu, sigma)
    
    # Plot the Gaussian curve
    plt.plot(x, pdf, 'r-', linewidth=2)

    # Set labels and title
    plt.xlabel(f'{column_name}')
    plt.ylabel('Probability Density')
    plt.title(fr'$\mathrm{{Histogram\ of\ {column_name}:}}\ \mu={mu:.3f},\ \sigma={sigma:.3f}$')
    plt.grid(True)
    
    # Show the plot
    plt.show()
# Plot histograms with smoother Gaussian curves for 'x', 'y', and 'z'
plot_histogram_with_gaussian(df[column_x], 'x')
plot_histogram_with_gaussian(df[column_y], 'y')
plot_histogram_with_gaussian(df[column_z], 'z')

    
```
